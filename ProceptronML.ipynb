#Problem Statement
Train your model to learn the table of 10

`y = 10*x`

##Data creation

```
xTrain = [0,1,2,3,4,5,....]

yTrain = [0,10,20,30,....]
```

````python
x = [i for i in range(21)] #List comprehension
print(xTrain)

````

````python
y = [i for i in range(10*20+1) if i%10==0]
print(y,end=" ")
````

##Approach 1

Defining the function for the eqn **y=10x**

````python
def tempFunc(x):
  y = 10*x
  return y

for value in x:
  print(tempFunc(value))
````

##Approach 2

Redundunt ML model

Needed things : 
**`xTrain`, `yTrain`,`xTest`,`yTest`**

````python
print(f'This is x: {x}')
print(f'This is y: {y}')
````

````python
xTrain = x[:-5]
yTrain = y[:-5]

xTest = x[-5:]
yTest = y[-5:]
print(f'''
Training Data 

xTrain : {xTrain}
yTrain : {yTrain}

Testing Data

xTest : {xTest}
yTest : {yTest}''')
````

````python
import tensorflow as tf
from tensorflow import keras
````

````python
#Perceptron Model

model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=1,input_shape=[1]) #units = no.of nurons && input_shape is shape
])
````

````python
model.compile(optimizer='adam',loss='mae')
````

````python
model.fit(x=xTrain,y=yTrain,validation_data=(xTest,yTest),epochs=5000)
````
